```python
from sqlalchemy import create_engine
import pandas as pd
import pymongo
```

```python
df = pd.read_csv("alumnos.csv")
df
```

```python
#Escribir a un csv
df.to_csv("datos/df/alumnos_coma.csv",index_label="ID")
df
```

```python
#Escribir a un csv
df.to_csv("datos/df/alumnos_pipe.csv",index=False, 
          sep = "|")

#! cat
#Muestra el contenido de un archivo
! type alumnos_pipe.csv 
```

```python
#Escribir un json
#En lista por campos
df.to_json('alumnos_campos.json',indent=2)
```

```python
#Escribir un json orientado a registros
#En lista por personas o cosas con sus descripciones
df.to_json('alumnos_registros.json',
           orient="records",
           indent=2)

#indent tabuladores de separaci√≥n
```

```python
#Escribir un json orientado a tablas
#En lista por *****
df.to_json('alumnos_tablas.json',
           orient="table",
           indent=2)
```

```python
#Escribir un json orientado a valores
#En lista solamente los datos de la bd
df.to_json('alumnos_valores.json',
           orient="values",
           indent=2)
```

```python
#Escribir en un excel
df.to_excel('datos/df/alumnos_excel.xlsx',
            sheet_name="AlumnosArcelia",
            index=False
            )

#Al estar en false, solo guarda los datos, sin contar la fila del indice
```

```python
#Escribir en multiples hojas en un archivo de excel
#df_2 = df.copy()
with pd.ExcelWriter('datos/df/alumnos_hojas.xlsx') as w:
    for i in range(5):
        hoja = f"AlumnosHoja{i}"
        df.to_excel(w, sheet_name=hoja, index=False)

#With abre un archivo y trabaja sobre el 
```
